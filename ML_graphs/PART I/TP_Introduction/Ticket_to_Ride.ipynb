##### 1. Implementing the underlying graph structure represented by the London game board <br>
--> The type of graph here is a multigraph

```python
G = nx.MultiDiGraph()
G.add_node(1, name="REGENT'S PARK")
G.add_node(2, name="BAKER STREET")
G.add_node(3, name="PICADILLY CIRCUS")
G.add_node(4, name="HYDE PARK")
G.add_node(5, name="BUCKINGHAM PALACE")
G.add_node(6, name="BIG BEN")
G.add_node(7, name="ELEPHANT & CASTLE")
G.add_node(8, name="TOWER OF LONDON")
G.add_node(9, name="BRICK LANE")
G.add_node(10, name="THE CHARTERHOUSE")
G.add_node(11, name="KING'S CROSS")
G.add_node(12, name="BRITISH MUSEUM")
G.add_node(13, name="COVENT GARDEN")
G.add_node(14, name="TRAFALGAR SQUARE")
G.add_node(15, name="GLOBE THEATRE")
G.add_node(16, name="WATERLOO")
G.add_node(17, name="ST PAUL'S")
print(f"Here are the {len(G)} nodes with their index:")
for _ in range(1, len(G)+1):
    print(_, "-->", G.nodes[_])
```

```python
G.add_edges_from([(1, 12, 3), (1, 2, 2), (1, 11, 3)]) # Regent's park
G.add_edges_from([(2, 4, 4), (2, 3, 4), (2, 12, 4)], color="red") # Baker street
G.add_edges_from([(3, 5, 2), (3, 14, 2), (3, 14, 2), (3, 12, 2), (3, 4, 2), (3, 4, 2), (3, 13, 1), (3, 13, 1)]) # Picadilly circus
G.add_edges_from([(4, 5, 1), (4, 5, 1)]) # Hyde Park
G.add_edges_from([(5, 6, 2), (5, 14, 2)]) # Buckingham palace
G.add_edges_from([(6, 14, 1), (6, 16, 1), (6, 7, 3)]) # Big ben
G.add_edges_from([(7, 16, 2), (7, 15, 2), (7, 8, 4)]) # Elephant & castle
G.add_edges_from([(8, 15, 3), (8, 17, 3), (8, 17, 3), (8, 17, 9)]) # tower of london
G.add_edges_from([(9, 17, 3), (9, 10, 3)]) # Brick lane 
G.add_edges_from([(10, 11, 3), (10, 12, 4)]) #the charterhouse
G.add_edges_from([(11, 12, 2)]) # king's cross
G.add_edges_from([(12, 13, 1), (12, 13, 1)]) # British museum
G.add_edges_from([(13, 14, 1), (13, 14, 1), (13, 17, 3), (13, 17, 3)]) # convent garden
G.add_edges_from([(14, 16, 2)]) # Trafalgar square
G.add_edges_from([(15, 17, 1), (15, 17, 1), (15, 16, 2)]) # Globe theatre
degree = dict(G.degree(weight="weight"))
```
